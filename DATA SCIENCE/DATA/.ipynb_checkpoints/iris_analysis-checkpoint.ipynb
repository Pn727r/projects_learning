```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.shape
```

```python
iris.columns
```

```python
iris.columns.isnull().sum()
```

```python
iris['species'].unique()
```

```python
p_width_data = iris.groupby('species')['petal_width'].mean()
p_length_data = iris.groupby('species')['petal_length'].mean()

print(p_width_data)
print(p_length_data)

#plt.figure(figsize=(8,8))
sns.scatterplot(x= iris.sepal_length , y= iris.sepal_width ,data=iris,  hue=iris.species, s=100 , legend=True)
plt.show()


#plt.figure(figsize=(8,8))
sns.scatterplot(x=iris.petal_length , y = iris.petal_width , data=iris , hue=iris.species , s=100 , legend=True)
plt.show()

```

```python
#plt.figure(figsize=(8,8))
sns.scatterplot(x= iris.sepal_length , y= iris.petal_length ,data=iris,  hue=iris.species, s=100 , legend=True)
plt.show()


#plt.figure(figsize=(8,8))
sns.scatterplot(x= iris.sepal_width , y= iris.petal_width ,data=iris,  hue=iris.species, s=100 , legend=True)
plt.show()
```

```python
petal = iris.groupby('species')[['petal_width', 'petal_length']].mean()
sepal = iris.groupby('species')[['sepal_width','sepal_length']].mean()

plt.ylabel('size in inch')
plt.plot(iris['species'].unique() , petal , 'o--r')
plt.plot(iris['species'].unique() ,sepal , 'D-b')
plt.legend(['pw','pl','sw','sl'])
plt.show()
```
