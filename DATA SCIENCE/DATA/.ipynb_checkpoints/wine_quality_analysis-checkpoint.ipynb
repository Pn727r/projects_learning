```python
'''
What is the highest rated quality of red wine, and how many wines have that rating?
What is the average citric acid content of the red wines with a quality rating of 8 or higher?
What is the correlation coefficient between the "volatile acidity" and "quality" columns?
'''
```

```python
corr_matrix = rwine[["volatile acidity" , "quality" , 'citric acid' , 'alcohol', 'density']] 

# Create a heatmap using seaborn
sns.heatmap(corr_matrix.corr(), annot=True, cmap="cool")
plt.show()
```

```python
'''
How many red wines have an alcohol content greater than 12.5% and a quality rating of 7 or higher?
'''
rwine.loc[(rwine.alcohol > 12.5 )& (rwine.quality  > 7)]
len(rwine.loc[(rwine.alcohol > 12.5 )& (rwine.quality  > 7) ])
```
