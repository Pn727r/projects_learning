# ploting sample point on map

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
day = np.arange(1,16)
temp = [36.6,37,37.7,39,40.1,43,43.4,45,45.6,40.1,44,45,46.8,47,47.8]
temp2 = [39,39.4,40,40.7,41,42.5,43.5,44,44.9,44,45,45.1,46,47,46]
style.use("ggplot")
```

```python
# plt.plot(day , temp , color='c' , marker = '.' , linestyle=':' , linewidth = 1 , markersize = 10)
plt.plot(day , temp ,'m.-' , linewidth = 1 , markersize = 10 , label = "Delhi Temperature point")  # same output as above
plt.plot(day , temp2 ,'y.-' , linewidth = 1 , markersize = 10 , label = "Mumbai Temperature point")  # same output as above

# plt.axis([0,15,0,50])
plt.title("delhi & mumbai temperature" , fontsize = 20)
plt.xlabel("DAYS" ,fontsize = 20)
plt.ylabel("Temperuture" , fontsize = 20)
plt.legend(loc = 4)
plt.grid(color='k')
plt.show()
```

# histo gram 

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
ml_stu_age = np.random.randint(18,45 , (100))
py_stu_age = np.random.randint(15,40,(100))
```

```python
plt.figure(figsize = (16,9))
style.use("ggplot")
plt.grid(color='m')
plt.hist([ml_stu_age , py_stu_age] , color= ['y','c'], bins = [15,20,25,30,35,40,45], rwidth=0.8 , orientation = 'vertical', label=["no. of ml_student", "no. of py_student"]) 
# plt.hist(py_stu_age , color= 'r', bins = [15,20,25,30,35,40,45], rwidth=0.8 , orientation = 'vertical', label="no. of py_student") 

plt.title("student histogram")
plt.xlabel("age cotegory" )
plt.ylabel("count of student")
plt.legend()
plt.show()
```

## Bar chart


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use("ggplot")
```

```python
classes = ["Python" , "R" ,"AI" , "ML" , "DataScience"]
classA_stu = [30,10,20,25,10]

classB_stu = [40,5,20,20,10]
classC_stu = [35,5,30,15,15]
```

```python
#plt.barh(classes , class1_stu , color='y')
plt.grid(color='k'  , alpha=0.2)
index = np.arange(len(classes))
width = 0.2
#plt.bar(index ,     class1_stu , color='c' , width=0.2 ,align='center',edgecolor='m' , linewidth=1 , alpha=1 ,linestyle='-' , label="class-1")
plt.bar(index ,     classA_stu , color='y' , width=0.2 ,align='center', linewidth=1 , alpha=1 ,linestyle='-' , label="class-A")
plt.bar(index+0.2 , classB_stu , color='m' , width=0.2 ,align='center', linewidth=1 , alpha=1 ,linestyle='-' , label="class-B")
plt.bar(index+0.4 , classC_stu , color='k' , width=0.2 ,align='center', linewidth=1 , alpha=1 ,linestyle='-' , label="class-C")

plt.xticks(index + width , classes)
plt.xlabel("Classes")
plt.title("Student per classes" , fontsize = 15)
plt.ylabel("No. Of Student" , fontsize = 15)
plt.legend()
plt.figure(figsize=(16,9))
plt.show()
```

# scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import style
style.use("ggplot")
```

```python
df = pd.read_csv('C:\\Users\\abc\\Desktop\\DATA SCIENCE\\DATA\\googleplaystore.csv' , nrows=20)
df.shape
df.head(4)
```

```python
x = df["Rating"]
y = df["Reviews"]
y2 = df["Installs"]
x.head(5)
y.head(5)
```

```python
plt.figure(figsize=(15,8))
plt.grid(color = 'k' , alpha=0.2)

plt.scatter(x , y , color='y' , marker="*" , s= 250 , alpha=1 , linewidth=1 , edgecolor="r")
plt.scatter(x , y2 , color='c' , marker="D" , s= 250 , alpha=0.3 , linewidth=1 )

plt.title("google play store apps plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
plt.show()
```

## PIE chart

```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python" , "R" ,"AI" , "ML" , "DataScience"]
classA_stu = [30,10,20,25,10]
```

```python
expld = [0.1,0,0.1,0,0]
colors = ['c','r','b','g','y']

textprops = {"fontsize" : 11 }
plt.pie(classA_stu , labels=classes , explode = expld, colors = colors ,autopct = "%0.2f%%" , radius = 1 , startangle= 0, textprops = textprops) # shadow= True
plt.show()
```

## SUB plot 

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
classes = ["Python","AI" , "ML" , "DataScience"]
classA_stu = [30,20,25,10]
classB_stu = [40,20,20,10]
classC_stu = [35,30,15,15]
classD_stu = [55,5,7,12]

expld = [0.1,0,0,0]
colors = ['c','r','b','g','y']

plt.subplot(3,3,1)
plt.pie(classA_stu , labels= classes , explode = expld , colors= colors  , autopct = "%0.1f%%" , radius = 2.5)

expld = [0,0.1,0,0]
plt.subplot(3,3,3)
plt.pie(classB_stu , labels= classes , explode = expld , colors= colors  , autopct = "%0.1f%%" , radius = 2.5)

expld = [0,0,0.1,0]
plt.subplot(3,3,7)
plt.pie(classC_stu , labels= classes , explode = expld , colors= colors  , autopct = "%0.1f%%" , radius = 2.5)

expld = [0,0,0,0.1]
plt.subplot(3,3,9)
plt.pie(classD_stu , labels= classes , explode = expld , colors= colors  , autopct = "%0.1f%%" , radius = 2.5)
plt.show()
```
