```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar  = [113,85,90,150,149,88,93,115,135,80,72,82,129]  # only one list or one array is enough in histogram
plt.hist(blood_sugar , bins= 4, rwidth=0.95 , color = 'red', histtype= 'step')
```

```python
blood_sugar_man   = [113,85,90,150,149,88,93,115,135,80,7,82,120]
blood_sugar_woman = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('Sugar range')
plt.ylabel('Total no of patients')
plt.title('Sugar Analysis')
plt.hist([blood_sugar_man , blood_sugar_woman], bins = [80,100,125,150] , rwidth = 0.80 , 
         color=['black' , 'purple'] , label=['man' , 'woman'] , orientation='vertical')
plt.legend()
```
