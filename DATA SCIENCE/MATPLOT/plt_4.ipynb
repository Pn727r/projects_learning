```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp = [1400,600,300,410,250]
things = ['home rent','food','internet bill','car exp','other']
plt.axis('equal')
plt.pie(exp , labels = things , radius = 1 , 
        autopct= '%0.2f%%' , shadow = False ,
        explode=[0.05,0.05,0.05,0.05,0.05] , startangle = 180) 
# plt.pie(exp , labels = things , radius = 1 , autopct= '%0.0f%%')
plt.show()
```
