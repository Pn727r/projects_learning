```python
missing_value = df.isnull().sum() / df.shape[0] * 100
missing_value
```

```python
missing_value_col = missing_value[missing_value > 20].keys()
missing_value_col
```

```python
lst = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
remove_data_column = df.drop(columns = lst)
remove_data_column.shape
```

```python
fresh = remove_data_column.select_dtypes(include = ['int64' , 'float64'])
fresh
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(fresh.isnull())
fresh[fresh.isnull().any(axis=1)]
fresh.isnull().sum()
```

```python
missing = [var for var in fresh.columns if fresh[var].isnull().sum() > 0]
missing
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i ,var in enumerate(missing):
    plt.subplot(2,2,i+1)
    sns.histplot(fresh[var] , bins= 20,kde_kws = {'linewidth' : 5 , 'color' : 'k'})
    
    
```

```python
# graph 2 = median Squide
# graph 3 = median squide
# graph 1 = mean symatric
# missing value is random that time this method required
final_mean = fresh.fillna(fresh.mean())
final_mean.isnull().sum().sum()

```

```python
plt.figure(figsize=(10,10))
sns.set()
for i ,var in enumerate(missing):
    plt.subplot(2,2,i+1)
    sns.histplot(fresh[var], bins= 20 ,color = 'c', label="original")
    sns.histplot(final_mean[var] , bins = 20 , color='r', label='means')
    plt.legend()
    
```

```python
final_median = fresh.fillna(fresh.mean())
final_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i ,var in enumerate(missing):
    plt.subplot(2,2,i+1)
    sns.histplot(fresh[var], bins= 20 ,color = 'c', label="original"  , kde_kws = {'linewidth' : 4 , 'color' : 'b'})
    sns.histplot(final_mean[var] , bins = 20 , color='b', label='means'  , kde_kws = {'linewidth' : 4 , 'color' : 'r'})
    sns.histplot(final_median[var] , bins = 20 , color='y', label='median'  , kde_kws = {'linewidth' : 4 , 'color' : 'c'})
    plt.legend()
```
