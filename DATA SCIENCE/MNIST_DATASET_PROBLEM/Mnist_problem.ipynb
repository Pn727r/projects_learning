```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras as kr
```

```python
(X_train , y_train) ,(X_test , y_test) = kr.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_train = X_train/255 #scale the high value to low value between 0 to 1
X_test = X_test/255 #scale the high value to low value between 0 to 1
X_train[0].shape
# 60000 row of data and each row contain 2d matrix like 28 row and 28 column
#  it represent as row column matrix in the form or number
```

```python
X_train_flattern = X_train.reshape(len(X_train) , 28*28)
X_test_flattern = X_test.reshape(len(X_test) , 28*28)
```

```python
X_train_flattern.shape
```

```python
X_test_flattern.shape
```

```python
X_train_flattern[0] # 784 number is a pixel representation
```

```python
X_test_flattern[0]
```

```python
model = kr.Sequential([kr.layers.Dense(10 , input_shape=(784 ,), activation = 'sigmoid')]) # stack of layer 10 is output labels
model.compile(
    optimizer = 'adam' ,
    loss = "sparse_categorical_crossentropy" ,
    metrics = ["accuracy"]
    )
model.fit(X_train_flattern ,y_train , epochs=500)
```

```python
model.evaluate(X_test_flattern , y_test)
```
