```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lst = [1,2,3,4,5,6,7]
lst2 = [50,53,45,48,49,46,35]
```

```python
plt.plot(lst,lst2)
```

```python
plt.xlabel('WEEK DAY')
plt.ylabel('TEMPERATURE')
plt.title("WEATHER")
plt.plot(lst,lst2 , color = 'red' , linewidth = 2, linestyle= 'dotted')
```

```python
plt.plot(lst,lst2 , '--r*')
plt.plot(lst,lst2 , 'rD--')
plt.plot(lst,lst2 , color='blue' , marker='D' , linestyle='--' , markersize = 7 , alpha=1)
```

```python
days = [1,2,3,4,5,6,7]
max_temp = [32,51,52,48,47,49,46]
min_temp = [43,42,40,44,33,35,37]
avg_temp = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.title('Weather')
plt.ylabel("Temp")
plt.plot(days,max_temp , label='max')
plt.plot(days,min_temp , label='min')
plt.plot(days,avg_temp , label='avg')
# plt.legend(loc='lower left')
plt.legend(loc='best' , shadow = True , fontsize = 'small')
plt.grid()
```
