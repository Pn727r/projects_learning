```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
data = pd.read_csv('home_price.csv')
data['area ']
data['price']
plt.xlabel('Area (sqr ft)')
plt.title("Price Prediction")
plt.ylabel('Price ($)')
plt.scatter(data['area '], data.price, color='r' , marker='D')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['area ']], data.price)
```

```python
pred_price  = reg.predict([[3300]])
print(f'HOME PRICE FOR {3300} sqr ft = ',round(pred_price[0],2))
```

```python
# print(reg.coef_)
# print(reg.intercept_)
print(f'{pred_price} = {3300}*{reg.coef_} + {reg.intercept_}')
print('y = mx + c' , 'ANS = ',(3300*reg.coef_) + reg.intercept_)
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.title('price prediction')
plt.scatter(data['area '], data.price , color='blue', marker='D')
plt.plot(data['area '], reg.predict(data[['area ']])  , color='red')
plt.show()
```
