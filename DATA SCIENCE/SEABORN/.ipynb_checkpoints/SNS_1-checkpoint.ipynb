# # heat map to find variable correlation 

```python
# --> correlation coefficient is statistical measure of the relationship between
# --> increase of decrease of x and y dependant variable
# --> it is between -1 , 0 , +1 

# correlation of x and y : R(xy) =                  sigma(i=1 , n) (Xi - X(avg))* (Yi - Y(avg)) 
#                                  ---------------------------------------------------------------------------
#                                  whole root of ((sigma(i=1, n) (Xi - X(avg))^2) * ( sigma(i=1, n) (Yi - Y(avg))^2))

```

```python
# method is pearson
plt.figure(figsize=(16,9))
ax = sns.heatmap(global_warm_data.corr() , annot=True , linewidth=3 )
ax.tick_params(size = 10 , color = 'w' , labelsize = 10 , labelcolor = 'w' )
plt.title("who is responsible for global warming" , fontsize = 25)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
cancer_data = load_breast_cancer()
cancer_data
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_data['data'],cancer_data['target']],
                        columns = np.append(cancer_data['feature_names'],['target']))
cancer_df.head(10)
```

```python
plt.figure(figsize=(25,25))
ax = sns.heatmap(cancer_df.corr() , annot=True , linewidth = 3 , )
ax.tick_params(size = 10 , color = 'w' , labelsize = 10 , labelcolor = 'w' )
plt.title("breast cancer heatmap" , fontsize = 25)
plt.show()
```
