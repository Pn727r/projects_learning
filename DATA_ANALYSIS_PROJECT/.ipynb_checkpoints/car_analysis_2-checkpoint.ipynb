```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car = pd.read_csv('car_data.csv')
```

```python
car.head(2)
```

# DATA CLEANING PROCESS

```python
car.isnull().sum()
```

## remove column have greater than 15% empty value 

```python
car.shape
check = 3742/11914 * 100 
check > 15
car.drop('Market Category' ,axis=1 , inplace=True)
```

```python
car.columns
```

```python
car.info()
```

## Get NaN or empty columns 

```python
miss_col = car.columns[car.isnull().any().tolist()]

#Engine Fuel Type        3
#Engine HP              69
#Engine Cylinders       30
#Number of Doors         6

fill = car['Engine Fuel Type'].mode()
car['Engine Fuel Type'] = car['Engine Fuel Type'].fillna(fill.values[0])
car['Engine Fuel Type'].isnull().sum()
```

```python
fill = car['Engine HP'].mode()
car['Engine HP'] = car['Engine HP'].fillna(fill.values[0])
car['Engine HP'].isnull().sum()
```

```python
fill = car['Engine Cylinders'].mode()
car['Engine Cylinders'] = car['Engine Cylinders'].fillna(fill.values[0])
car['Engine Cylinders'].isnull().sum()
```

```python
fill = car['Number of Doors'].mode()
car['Number of Doors'] = car['Number of Doors'].fillna(fill.values[0])
car['Number of Doors'].isnull().sum()
```

```python
car.isnull().sum()
```

# Data Manipulation


## What is the distribution of car prices in the dataset? 

```python
plt.hist(car.MSRP , bins=50 , color='r')
plt.xlabel('Car price')
plt.ylabel('Number of cars')
plt.title('Distribution of car prices')
plt.show()
```


# What is the most popular car make and model in the dataset? 

```python
c = car.groupby(['Make' , 'Model']).size().reset_index(name='counts') 
c.sort_values(by="counts" , ascending = False).head(7)
```



## What is the distribution of car mileage in the dataset? Are there any outliers?

```python
plt.hist((car['city mpg']+car['highway MPG'])/2 , bins=50) 
plt.xlabel('Mileage')
plt.ylabel('Number of cars')
plt.title('Distribution of car mileage')
plt.show()

z = car['city mpg']+car['highway MPG']/2    
outlier = (z - z.mean())/ z.std()
(outlier > 3).sum()
# 110 outlier detected 
```

## What is the average city and highway miles per gallon (MPG) for different car makes?
 

```python
plt.figure(figsize=(10,4))
mpg = car.groupby('Make')[['city mpg', 'highway MPG']].mean()
plt.plot(mpg.index , mpg['city mpg'] ,color='r' , marker='v')
plt.bar(mpg.index , mpg['city mpg'] ,color='y')
plt.title('Average MPG based on car company')
plt.ylabel('Average MPG')
plt.xticks(rotation=90)
plt.show()
```

## Which car make and model has the highest horsepower in the dataset?


```python
plt.figure(figsize=(6,12))
car.columns
values = car.groupby(['Make', 'Model'])['Engine HP'].max().reset_index(name = 'max')
new = values.sort_values(by='max' , ascending = False )
plt.barh(new.groupby('Make')['max'].max().index , new.groupby('Make')['max'].max().values , color='g')
plt.xlabel('Car Company')
plt.title('Highest horse power covered by each company')
plt.ylabel('Engine HorsePower')
plt.show()
```

## How many cars in the dataset have manual vs. automatic transmission?
 

```python
#fill = car['Transmission Type'].mode()
#car['Transmission Type'] = car['Transmission Type'].replace('UNKNOWN',fill.values[0])
count = car.groupby('Transmission Type').size()
plt.pie(count, autopct='%1.1f%%' , labels=count.index)
plt.show()
count
```



## How many cars in the dataset have four-wheel drive or all-wheel drive?



## What is the most common type of fuel used by cars in the dataset? 

```python
fuel = car.groupby('Engine Fuel Type').size()
plt.barh(fuel.index , fuel.values)
plt.show()

# regular unleaded fuel is common 
```



## How many model for each make have large,compact , midsize car ?

```python
plt.figure(figsize=(8,4))
x = car[car['Vehicle Size'] == 'Large'][['Make' , 'Model']].groupby('Make')['Model'].size()
y = car[car['Vehicle Size'] == 'Midsize'][['Make' , 'Model']].groupby('Make')['Model'].size()
z = car[car['Vehicle Size'] == 'Compact'][['Make' , 'Model']].groupby('Make')['Model'].size()


plt.xticks(rotation = 90)
plt.title('No of Large size car for each company')
plt.xlabel('car company')
plt.ylabel('No of car have large size')
plt.bar(x.index , x.values , color='r' , align="center")
plt.show()

plt.figure(figsize=(8,4))
plt.xticks(rotation = 90)
plt.title('No of Midsize car for each company')
plt.xlabel('car company')
plt.ylabel('No of car have Midsize')
plt.bar(y.index , y.values , color='b' , align="center")
plt.show()


plt.figure(figsize=(8,4))
plt.xticks(rotation = 90)
plt.title('No of Compact size car for each company')
plt.xlabel('car company')
plt.ylabel('No of car have Compact size')
plt.bar(z.index , z.values , color='g' , align="center")
plt.show()
```

## correlations between numerical values

```python
car.select_dtypes(include=['float64','int64']).head(5)
```

```python
relation = car[[
    'Engine HP',
    'Engine Cylinders',
    'Number of Doors',
    'highway MPG',
    'city mpg',
    'Popularity','MSRP',]] 
sns.heatmap(relation.corr() , cmap='brg' , annot=True)
plt.show()
```

                     conclusion1 : MSRP 66% increase with Engine HP increase
                     conclusion2 : MSRP 53% increase with Engine Cylinders increase 
                     conclusion3 : Engine HP 77% increase with Engine Cylinders increase  
                     conclusion4 : Engine HP 37% and 36% decrease with Highway or City MPG increase
                     conclusion5 : Engine Cylinders 61% and 57% decrease with Highway or City MPG increase
