```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.shape
#iris.to_csv('iris_data.csv' , index=False)
```

```python
iris.columns
```

## Are there any missing values in the Iris dataset?

```python
iris.columns.isnull().sum()
```

```python
iris['species'].unique()
```

## What is the average petal width of each species of iris?

```python
plt.figure(figsize=(4,3))
plt.bar(iris.groupby('species')['petal_length'].mean().index , iris.groupby('species')['petal_length'].mean().values)
plt.ylabel('petal lenght (cm)')
plt.show()
```

## What is the average sepal length of each species of iris?

```python
plt.figure(figsize=(4,3))
plt.bar(iris.groupby('species')['sepal_length'].mean().index , iris.groupby('species')['sepal_length'].mean().values)
plt.ylabel('sepal lenght (cm)')
plt.show()
```

## What is the median sepal width of the Iris dataset?


```python
plt.figure(figsize=(4,3))
plt.bar(iris.groupby('species')['sepal_width'].median().index , iris.groupby('species')['sepal_width'].median().values)
plt.ylabel('sepal width (cm)')
plt.show()
```

## What is the range of sepal width in the Iris dataset?


```python
iris.sepal_width.max() # 4.4
iris.sepal_width.min() # 2
# range is 2 to 4.4
```

## What is the range of petal length in the Iris dataset?


```python
iris.petal_length.max() # 6.9
iris.petal_length.min() # 1.0
# range is 1.0 to 6.9
```

## Correlation between numerical columns

```python
numerical = iris.select_dtypes(include = 'float64')
sns.heatmap(numerical.corr() , cmap='cool' , annot=True)
plt.show()
```

# What is the distribution of sepal length in the Iris dataset?

```python
sns.histplot(data=iris,x='sepal_length' , bins=10 , color='c')
plt.show()
```

## How does sepal length vary between the different species of iris in the dataset?
## How does sepal width vary between the different species of iris in the dataset?

```python
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Variation in Sepal Length between Different Species')
plt.show()

sns.boxplot(data=iris, x='species', y='sepal_width')
plt.title('Variation in Sepal Width between Different Species')
plt.show()
```

## How does the distribution of petal length compare between the different species of iris in the dataset?

## How does the distribution of petal width compare between the different species of iris in the dataset?

```python
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title('Variation in Petal Length between Different Species')
plt.show()

sns.boxplot(data=iris, x='species', y='petal_width')
plt.title('Variation in Petal Width between Different Species')
plt.show()
```

## Are there any outliers in the petal length column of the setosa in dataset using mathematical way ?


```python
iris_setosa = iris[iris.species == 'setosa'] 
q1 ,q2 ,q3 = np.percentile(iris_setosa['petal_length'] , [25,50,75] ,axis=0)  
IQR = q3 - q1
lower = q1 - (1.5 * IQR) # 1.1374
higher = q3 + (1.5 * IQR) # 1.8375
iris_setosa[(iris_setosa.petal_length < lower) | (iris_setosa.petal_length > higher)]
```
