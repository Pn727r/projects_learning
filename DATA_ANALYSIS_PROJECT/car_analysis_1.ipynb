```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cd = pd.read_csv('Cars Data.csv')
```

```python
cd.head()
```

```python
cd.company_name.unique()
```

```python
cd.isnull().sum()
```

```python
cd.info()
```

```python
# drop nan value in object columns 
cd.dropna(subset=['company_name' , 'Model' , 'Type' ,'Origin' , 'DriveTrain' , 'MSRP' , 'Invoice'] , inplace = True)
```

```python
cd.isnull().sum()
```

```python
# clean data
cd.Cylinders = cd['Cylinders'].replace("missing",np.nan)
cd.Cylinders.fillna(cd.Cylinders.mean() , inplace = True)
cd.Cylinders.isnull().sum()
```

```python
# check each car company how many row containt it self
cd.company_name.value_counts()
```

```python
# show all record which belongs to asia or europe
#cd.loc[(cd.Origin == "Asia") | (cd.Origin == 'Europe')]
cd[cd.Origin.isin(['Asia' , 'Europe'])]
```

```python
# remove all the records where weight above the 4000
cd[~(cd.Weight > 4000)]
```

```python
# increase all the vlaue of MPG_City columns by 3 
cd.MPG_City = cd.MPG_City + 3
cd.head()
```

```python
# replace $ sign
cd.MSRP = cd.MSRP.str.replace(r'$' , "" , regex=True )
cd.MSRP = cd.MSRP.str.replace(r',' , "" , regex=True ).astype(float)
cd.Invoice = cd.Invoice.str.replace(r'$' , "" , regex=True)
cd.Invoice  = cd.Invoice.str.replace(r',',"",regex=True).astype(float)
```

```python
# car model which contain horsepower > 400
cd[cd.Horsepower > 400][['company_name' , 'Model']]
```

```python
category =cd.groupby('company_name')['MSRP'].mean()
plt.figure(figsize=(8,8))
plt.xticks(rotation=90)
plt.title('Average value of car for each company')
plt.bar(category.index , category.values , color='r')
```

```python
# relation between variables
relation = cd[['EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight']]
sns.heatmap(relation.corr(), annot = True , cmap="cool")
```

```python
# total number of model for each company
no_of_model = cd.groupby('company_name')['Model'].count()
plt.figure(figsize=(8,8))
plt.xticks(rotation=90)
plt.title('No of car models company makes')
plt.bar(no_of_model.index , no_of_model.values , color='c')
plt.show()
```
